## Pandas Profiling

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('Travel.csv')
```

```python
df.head()
```

```python
## To create simple report quickly
profile = ProfileReport(df,title='Travel Dataset Report',explorative=True)
```

```python
profile.to_widgets()
```

```python
profile.to_file('travel.html')
```

## Autoviz

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
filename = "Travel.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
